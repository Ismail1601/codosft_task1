```python
from sklearn.model_selection import train_test_split
```

```python
import pandas as pd
titanic = pd.read_csv('Titanic-Dataset.csv')
titanic['Age'].fillna(titanic['Age'].median(), inplace=True)
titanic.drop(columns=['Cabin'], inplace=True)
titanic['Embarked'].fillna(titanic['Embarked'].mode()[0], inplace=True)
titanic['Sex'] = titanic['Sex'].map({'male': 0, 'female': 1})
titanic = pd.get_dummies(titanic, columns=['Embarked'], drop_first=True)
```

```python
x = titanic.drop(columns=['PassengerId', 'Name', 'Ticket', 'Survived'])
y = titanic['Survived']
```

```python
print(f'Length of x: {len(x)}')
```

```python
print(f'Length of y: {len(y)}')
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
print(f'Length of x_train: {len(x_train)}')
```

```python
print(f'Length of y_train: {len(y_train)}')
```

```python
print(f'Length of x_test: {len(x_test)}')
```

```python
print(f'Length of y_test: {len(y_test)}')
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
log = RandomForestClassifier(random_state=42)
log.fit(x_train, y_train)
```

```python
pred = log.predict(x_test)
print(pred)
```

```python
accuracy = accuracy_score(y_test, pred)
classification_report_result = classification_report(y_test, pred)
confusion_matrix_result = confusion_matrix(y_test, pred)
```

```python
print(f'Accuracy: {accuracy}')
```

```python
print('Classification Report:')
print(classification_report_result)
```

```python
print('Confusion Matrix:')
print(confusion_matrix_result)
```
